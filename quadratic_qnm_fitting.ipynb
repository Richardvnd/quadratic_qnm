### Initialisation

```python
import numpy as np
import matplotlib.pyplot as plt

import qnmfits

sim = qnmfits.SXS(ID=305, zero_time=(2,2))
```

### Fit NR data to simple linear model 

Modes chosen to replicate those in package_tutorial. 

Most general list of modes: 

modes = [(l,m,n,p) for l in range(2,L) for m in range(-l,l) for n in range(0,N) for p in [-1,1]]


```python
modes = [(l,m,n,1) for l in range(2,4) for m in range(2,3) for n in range(0,8)]

best_fit = qnmfits.multimode_ringdown_fit(
    sim.times,
    sim.h,
    modes,
    Mf=sim.Mf,
    chif=sim.chif_mag,
    t0=0,
)
    
qnmfits.plot_mode_amplitudes(
    best_fit['C'], best_fit['mode_labels'], log=False
    )
```

# Create a new model that includes quadratic terms

For now, just include (2201)x(2201) mode

```python
new_modes = modes + [(2,2,0,1,2,2,0,1)]

best_fit = qnmfits.multimode_ringdown_fit(
    sim.times,
    sim.h,
    new_modes,
    Mf=sim.Mf,
    chif=sim.chif_mag,
    t0=0,
)

print(f"Mismatch = {best_fit['mismatch']}")
```
